```python
import numpy as np
import pandas as pd
import re
import json
from nltk.corpus import stopwords
import nltk
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import confusion_matrix, classification_report
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense, Dropout, Flatten
import matplotlib.pyplot as plt
import seaborn as sns

# Initialize stopwords
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))

# Preprocessing function
def preprocess_text(text):
    text = text.lower()
    text = re.sub(r'[^\w\s]', '', text)  # Remove punctuation
    text = re.sub(r'\d+', '', text)      # Remove numbers
    text = ' '.join([word for word in text.split() if word not in stop_words])  # Remove stopwords
    return text

# Load Bible translations
columns = ['verse_id', 'book_id', 'chapter', 'verse', 'text']
translations = ['t_kjv.csv', 't_asv.csv', 't_bbe.csv', 't_dby.csv', 't_wbt.csv', 't_web.csv', 't_ylt.csv']
translation_names = ['KJV', 'ASV', 'BBE', 'DBY', 'WBT', 'WEB', 'YLT']

all_translations = pd.DataFrame()
for file, name in zip(translations, translation_names):
    file_path = f'bible_data/bible_databases-master/bible_databases-master/csv/{file}'
    df = pd.read_csv(file_path)
    df.columns = columns
    df['translation'] = name
    df['cleaned_text'] = df['text'].apply(preprocess_text)
    all_translations = pd.concat([all_translations, df], ignore_index=True)

print(f"All translations loaded. Total rows: {len(all_translations)}")

# Merge with book metadata
key_english = pd.read_csv('bible_data/bible_databases-master/bible_databases-master/csv/key_english.csv')
key_english.columns = ['book_id', 'book_name', 'testament', 'genre']
all_translations = all_translations.merge(key_english, on='book_id', how='left')
print(f"After merging with metadata. Total rows: {len(all_translations)}")

# Map authors to verses using the provided author list
author_list = {
    "Genesis": "Moses",
    "Exodus": "Moses",
    "Leviticus": "Moses",
    "Numbers": "Moses",
    "Deuteronomy": "Moses",
    "Joshua": "Joshua",
    "Judges": "Samuel, Nathan, Gad",
    "Ruth": "Samuel, Nathan, Gad",
    "1 Samuel (1 Kings)": "Samuel, Nathan, Gad",
    "2 Samuel (2 Kings)": "Samuel, Nathan, Gad",
    "1 Kings (3 Kings)": "Jeremiah",
    "2 Kings (4 Kings)": "Jeremiah",
    "1 Chronicles": "Ezra",
    "2 Chronicles": "Ezra",
    "Ezra": "Ezra",
    "Nehemiah": "Nehemiah, Ezra",
    "Esther": "Mordecai",
    "Job": "Job,Moses",
    "Psalms": "David,Asaph, Ezra, the sons of Korah, Heman, Ethan, Moses",
    "Proverbs": "Solomon ,Agur(30) and Lemuel(31)",
    "Ecclesiastes": "Solomon",
    "Song of Solomon (Canticles)": "Solomon",
    "Isaiah": "Isaiah",
    "Jeremiah": "Jeremiah",
    "Lamentations": "Jeremiah",
    "Ezekiel": "Ezekiel",
    "Daniel": "Daniel",
    "Hosea": "Hosea",
    "Joel": "Joel",
    "Amos": "Amos",
    "Obadiah": "Obadiah",
    "Jonah": "Jonah",
    "Micah": "Micah",
    "Nahum": "Nahum",
    "Habakkuk": "Habakkuk",
    "Zephaniah": "Zephaniah",
    "Haggai": "Haggai",
    "Zechariah": " Zechariah",
    "Malachi": "Malachi",
    "Matthew": "Matthew",
    "Mark": "John Mark",
    "Luke": "Luke",
    "John": "John, the Apostle",
    "Acts": "Luke",
    "Romans": "Paul",
    "1 Corinthians": "Paul",
    "2 Corinthians": "Paul",
    "Galatians": "Paul",
    "Ephesians": "Paul",
    "Philippians": "Paul",
    "Colossians": "Paul",
    "1 Thessalonians": "Paul",
    "2 Thessalonians": "Paul",
    "1 Timothy": "Paul",
    "2 Timothy": "Paul",
    "Titus": "Paul",
    "Philemon": "Paul",
    "Hebrews": "Paul, Luke, Barnabas, Apollos",
    "James": "James the brother of Jesus and Jude (not the Apostle, brother of John).",
    "1 Peter": "Peter",
    "2 Peter": "Peter",
    "1 John": "John, the Apostle",
    "2 John": "John, the Apostle",
    "3 John": "John, the Apostle",
    "Jude": "Jude, the brother of Jesus",
    "Revelation": "John, the Apostle"
}
all_translations['author'] = all_translations['book_name'].map(author_list)
all_translations = all_translations.dropna(subset=['author'])
print(f"After mapping authors. Total rows: {len(all_translations)}")

# Tokenize text
tokenizer = Tokenizer(num_words=20000)
tokenizer.fit_on_texts(all_translations['cleaned_text'])
sequences = tokenizer.texts_to_sequences(all_translations['cleaned_text'])
x_data = pad_sequences(sequences, maxlen=100)

# Encode labels
label_encoder_book = LabelEncoder()
label_encoder_author = LabelEncoder()
label_encoder_translation = LabelEncoder()

y_data_book = label_encoder_book.fit_transform(all_translations['book_name'])
y_data_author = label_encoder_author.fit_transform(all_translations['author'])
y_data_translation = label_encoder_translation.fit_transform(all_translations['translation'])

# Debugging Label Encoding
print("Label encoding completed.")
print(f"Number of unique books: {len(label_encoder_book.classes_)}")
print(f"Number of unique authors: {len(label_encoder_author.classes_)}")
print(f"Number of unique translations: {len(label_encoder_translation.classes_)}")

# Split data into training and testing sets
x_train, x_test, y_train_book, y_test_book = train_test_split(x_data, y_data_book, test_size=0.2, random_state=42)
_, _, y_train_author, y_test_author = train_test_split(x_data, y_data_author, test_size=0.2, random_state=42)
_, _, y_train_translation, y_test_translation = train_test_split(x_data, y_data_translation, test_size=0.2, random_state=42)

assert len(x_train) == len(y_train_book), "Mismatch between x_train and y_train_book sizes."
assert len(x_test) == len(y_test_book), "Mismatch between x_test and y_test_book sizes."

def build_and_train_model(model_type, x_train, y_train, x_test, y_test, num_classes, task_name, top_n=10):
    print(f"\nBuilding {model_type} model for {task_name} classification...")
    print(f"x_train size: {len(x_train)}, y_train size: {len(y_train)}")
    print(f"x_test size: {len(x_test)}, y_test size: {len(y_test)}")
    print(f"Number of classes: {num_classes}")

    # Define the model
    if model_type == "CNN":
        model = Sequential([
            Embedding(input_dim=20000, output_dim=128, input_length=100),  # Output: (batch_size, 100, 128)
            Dense(128, activation='relu'),  # Output: (batch_size, 100, 128)
            Dropout(0.5),
            Dense(64, activation='relu'),  # Output: (batch_size, 100, 64)
            Dropout(0.5),
            Flatten(),  # Flatten to (batch_size, 6400) for Dense layer
            Dense(num_classes, activation='softmax')  # Output: (batch_size, num_classes)
        ])
    elif model_type == "RNN":
        model = Sequential([
            Embedding(input_dim=20000, output_dim=128, input_length=100),  # Output: (batch_size, 100, 128)
            LSTM(128, return_sequences=False),  # Output: (batch_size, 128)
            Dropout(0.5),
            Dense(64, activation='relu'),  # Output: (batch_size, 64)
            Dropout(0.5),
            Dense(num_classes, activation='softmax')  # Output: (batch_size, num_classes)
        ])
    elif model_type == "RCNN":
        model = Sequential([
            Embedding(input_dim=20000, output_dim=128, input_length=100),  # Output: (batch_size, 100, 128)
            LSTM(128, return_sequences=True),  # Output: (batch_size, 100, 128)
            Dropout(0.5),
            Flatten(),  # Flatten to (batch_size, 12800) for Dense layer
            Dense(64, activation='relu'),  # Output: (batch_size, 64)
            Dropout(0.5),
            Dense(num_classes, activation='softmax')  # Output: (batch_size, num_classes)
        ])
    else:
        raise ValueError("Invalid model type. Choose from 'CNN', 'RNN', or 'RCNN'.")

    print(f"Model Summary for {model_type} - {task_name}:")
    model.summary()  # Print the model architecture for debugging

    # Compile the model
    model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

    # Train the model
    print(f"Training {model_type} model for {task_name}...")
    history = model.fit(x_train, y_train, epochs=10, batch_size=32, validation_split=0.2)
    loss, accuracy = model.evaluate(x_test, y_test)
    print(f"Test Accuracy ({task_name} - {model_type}): {accuracy}")

    # Save the model
    model_path_keras = f"{task_name.lower()}_{model_type.lower()}.keras"
    model.save(model_path_keras)
    print(f"Model saved as: {model_path_keras}")
    
    # Save accuracy/loss data
    history_data = {
        "accuracy": history.history["accuracy"],
        "val_accuracy": history.history["val_accuracy"],
        "loss": history.history["loss"],
        "val_loss": history.history["val_loss"]
    }
    history_path = f"{task_name.lower()}_{model_type.lower()}_history.json"
    with open(history_path, "w") as f:
        json.dump(history_data, f)
    print(f"Training history saved as: {history_path}")

    # Predictions
    y_pred = model.predict(x_test)
    y_pred_classes = np.argmax(y_pred, axis=1)

    # Confusion matrix logic
    cm_full = confusion_matrix(y_test, y_pred_classes)

    # Select the appropriate label encoder
    label_encoder = {
        "Book": label_encoder_book,
        "Author": label_encoder_author,
        "Translation": label_encoder_translation
    }[task_name]

    # Create confusion matrix DataFrame
    cm_df = pd.DataFrame(cm_full, index=label_encoder.classes_, columns=label_encoder.classes_)
    top_classes = cm_df.sum(axis=1).sort_values(ascending=False).head(top_n).index
    cm = cm_df.loc[top_classes, top_classes].values

    # Plot and save confusion matrix
    plt.figure(figsize=(12, 10))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=top_classes, yticklabels=top_classes)
    plt.title(f"Confusion Matrix - {task_name} ({model_type})")
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.savefig(f"confusion_matrix_{task_name.lower()}_{model_type.lower()}.png")
    print(f"Saved confusion matrix for {task_name} ({model_type}) as PNG.")
    plt.show()

    # Call after each model training
    plot_training_curves(history, model_type, task_name)

    # Classification report
    print(f"\nClassification Report ({task_name} - {model_type}):")
    print(classification_report(y_test, y_pred_classes, target_names=label_encoder.classes_, zero_division=0))

    return accuracy, history

# Plot and save accuracy/loss curves
def plot_training_curves(history, model_type, task_name):
    plt.figure(figsize=(12, 5))
    
    # Accuracy plot
    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'], label='Training Accuracy')
    plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
    plt.title(f'{model_type} - {task_name} Accuracy')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.legend()

    # Loss plot
    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'], label='Training Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.title(f'{model_type} - {task_name} Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()

    # Save the plot
    plt.tight_layout()
    plt.savefig(f'{model_type.lower()}_{task_name.lower()}_accuracy_loss.png')
    plt.show()

results = {}
for model_type in ["RNN", "CNN", "RCNN"]:
    for task_name, y_train, y_test, num_classes in [
        ("Book", y_train_book, y_test_book, len(label_encoder_book.classes_)),
        ("Author", y_train_author, y_test_author, len(label_encoder_author.classes_)),
        ("Translation", y_train_translation, y_test_translation, len(label_encoder_translation.classes_))
    ]:
        accuracy, _ = build_and_train_model(model_type, x_train, y_train, x_test, y_test, num_classes, task_name)
        results[f"{model_type}_{task_name}"] = accuracy

# Save results as a JSON file
results_path = "model_accuracies.json"
with open(results_path, "w") as f:
    json.dump(results, f)
print(f"Model accuracies saved as: {results_path}")
    
    
def plot_combined_accuracy_loss(model_type, tasks=["Book", "Author", "Translation"]):
    """
    Plot combined accuracy and loss curves for all tasks in one figure.
    
    Args:
    - model_type (str): The type of model (e.g., RNN, CNN, RCNN).
    - tasks (list): The classification tasks (default: ["Book", "Author", "Translation"]).
    
    Generates:
    - Combined accuracy/loss PNG.
    """
    plt.figure(figsize=(12, 10))
    
    # Accuracy Subplot
    plt.subplot(2, 1, 1)
    for task_name in tasks:
        history_path = f"{task_name.lower()}_{model_type.lower()}_history.json"
        with open(history_path, "r") as f:
            history = json.load(f)
        plt.plot(history["accuracy"], label=f"{task_name} - Train")
        plt.plot(history["val_accuracy"], linestyle="--", label=f"{task_name} - Val")
    plt.title(f"{model_type} Accuracy (All Tasks)")
    plt.xlabel("Epoch")
    plt.ylabel("Accuracy")
    plt.legend()
    plt.grid()

    # Loss Subplot
    plt.subplot(2, 1, 2)
    for task_name in tasks:
        history_path = f"{task_name.lower()}_{model_type.lower()}_history.json"
        with open(history_path, "r") as f:
            history = json.load(f)
        plt.plot(history["loss"], label=f"{task_name} - Train")
        plt.plot(history["val_loss"], linestyle="--", label=f"{task_name} - Val")
    plt.title(f"{model_type} Loss (All Tasks)")
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.legend()
    plt.grid()

    # Save and Show the Figure
    plt.tight_layout()
    combined_plot_path = f"{model_type.lower()}_combined_accuracy_loss.png"
    plt.savefig(combined_plot_path)
    print(f"Combined accuracy/loss plots saved as: {combined_plot_path}")
    plt.show()

# Generate combined plots for each model type
for model_type in ["RNN", "CNN", "RCNN"]:
    plot_combined_accuracy_loss(model_type)
```
